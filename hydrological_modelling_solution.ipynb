```python
import matplotlib.pyplot as plt
```

```python
x = [0, 0.05, 0.1, 0.2, 0.4, 0.8, 1.6]
y = [0, 0.1, 12.3, 30.1, 77.9, 91.8, 96.2]
plt.figure(figsize=(15, 7))
plt.plot(x, y)
plt.vlines(0.09, 0, 10, linestyle="dashed")
plt.hlines(10, 0, 0.09, linestyle="dashed")
plt.vlines(0.125, 0, 17, linestyle="dashed")
plt.hlines(17, 0, 0.125, linestyle="dashed")
plt.title('Grain-size composition curve', fontsize = 16)
plt.xlabel('d (mm)', fontsize = 16)
plt.ylabel('p (%)', fontsize = 16)
plt.xticks([0.1, 0.2, 0.4, 0.8, 1.6], [0.1, 0.2, 0.4, 0.8, 1.6], fontsize = 16)
plt.yticks([10, 17, 25, 50, 75, 100], [10, 17, 25, 50, 75, 100], fontsize = 16)
plt.xlim(0,None)
plt.ylim(0,None)
plt.gca().get_yticklabels()[0].set_color("red")
plt.gca().get_yticklabels()[1].set_color("red")
```

```python
def kfunc(g, v, n, deff):
    return 8.3*(10**-3)*g/v*(n**3)/((1-n)**2)*(deff**2)

def viscosity(T, A = 2.414 * (10 ** -5), B = 247.8, C = 140):
    return A * 10 ** (B / (T + 273.15 - C))
```

```python
g = 9.8 #m/s
n = 0.35 #?
T = 15 #C
deff = 0.07 #mm

v = viscosity(T)
k = kfunc(g, v, n, deff)
print(viscosity(T))
print(kfunc(g, v, n, deff))
```

```python
g = 9.8 #m/s
deff = 0.07 #mm
T = 15 #C
narray = [0.25, 0.3, 0.35, 0.4]
v = viscosity(T)
k = []
for n in narray:
    k.append(kfunc(g, v, n, deff))
```

```python
plt.figure(figsize=(15, 7))
plt.plot(narray, k)
plt.title('K vs n dependency', fontsize = 16)
plt.xlabel('n', fontsize = 16)
plt.ylabel('K', fontsize = 16)
plt.xticks([0.25, 0.3, 0.35, 0.4], [0.25, 0.3, 0.35, 0.4], fontsize = 16)
plt.yticks(fontsize = 16)
```

```python
g = 9.8 #m/s
deff = 0.07 #mm
Tarray = [10, 15, 20, 25, 30, 35, 40, 45, 50] #C
n = 0.35
k = []
for T in Tarray:
    v = viscosity(T)
    k.append(kfunc(g, v, n, deff))
```

```python
plt.figure(figsize=(15, 7))
plt.plot(Tarray, k)
plt.title('K vs T dependency', fontsize = 16)
plt.xlabel('T', fontsize = 16)
plt.ylabel('K', fontsize = 16)
plt.xticks([10, 15, 20, 25, 30, 35, 40, 45, 50], [10, 15, 20, 25, 30, 35, 40, 45, 50], fontsize = 16)
plt.yticks(fontsize = 16)
```
